```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Preprocessed data.csv', parse_dates= True, index_col = 'date_time', low_memory = False )
```

```python
df.head()
```

```python
df_df=df.copy(deep=True)
df_df.head()
```

```python
df_df.reset_index(inplace=True)
df_df.head()
```

```python
def month(x):
    return x.month
df_df['Month']= df_df['date_time'].apply(lambda x: month(x))
def year(x):
    return x.year
df_df['Year']=df_df['date_time'].apply(lambda x: year(x))
def hour(x):
    return x.hour
df_df['Hour']= df_df['date_time'].apply(lambda x: hour(x))
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.barplot(x="Year", y="Global_active_power",data=df_df)
```

```python


plt.figure(figsize=(10,6))
sns.lineplot(x="Year", y="Global_active_power",data=df_df)
```

```python

plt.figure(figsize=(10,6))
sns.lineplot(x="Month", y="Global_active_power",data=df_df)
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Month", y="Global_active_power",data=df_df)
```
